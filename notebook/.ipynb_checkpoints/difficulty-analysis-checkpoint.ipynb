```python
import pandas as pd 
import numpy 
import RNA 
import matplotlib.pyplot as plt 
import subprocess
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os 
import pickle
```

```python
df = pd.read_csv("sorted-full-eterna-dt.csv")
eterna_strc = []
for line in df.values: 
    eterna_strc.append([line[-1].strip(), line[1]])
    
lengths = []
bp_density = []
all_maker = []
for line in df.values : 
    lengths.append(len(line[-1]))
    list_ = list(line[-1])
    bp_density.append(list_.count("(")/float(len(line[-1])))
    all_maker.append(line[-2])

plt.scatter(lengths, bp_density,c="r" )
plt.title("Eterna dataset")
plt.xlabel("Length of target")
plt.ylabel("bp_density")
plt.show()
```

```python
def get_bp_position(structure) : 
    position = RNA.ptable(structure)
    position = list(position[1:])

    base_paire_pos = []
    for i in range(len(position)) :
        if position[i] != 0 :
            if (position[i]-1,i) in base_paire_pos : 
                continue; 
            else : 
                base_paire_pos.append((i,position[i]-1))

    return base_paire_pos

def getInteriorCoord(sequence, target) : 
    cmd_input = str(sequence)+"\n"+str(target)+"\n"
    os.system("echo '"+cmd_input+"' | RNAeval -v | grep Interior\ loop | tr -d A-Z-a-z-'('-')'|cut -d ':' -f 1 > interior.loop")
    
    with open("interior.loop", "r") as loops : 
        data_r = loops.read().split(";")
        list_ = []
        for dt in data_r : 
            list_+=filter(None,dt.strip().split("\n"))
    return numpy.array(list(set([tuple(map(int,string.strip().split(","))) for string in list_]))) -1

def getHairepinCoord(sequence, target) : 
    cmd_input = str(sequence)+"\n"+str(target)+"\n"
    #print cmd_input 
    os.system("echo '"+cmd_input+"'| RNAeval -v |grep Hairpin | tr -d A-Z-a-z-'('-')' | cut -d ':' -f 1 > hairpin.loop")
    
    hairpins = []
    with open("hairpin.loop", "r") as loops : 
        data_r = loops.read().split("\n")
        for dt in data_r : 
            if dt != '' : 
                #print map(int, dt.strip().split(","))
                hairpins.append(tuple(numpy.array(map(int, dt.strip().split(',')))-1))
    return hairpins 

def getMultiCoord(sequence, target) : 
    cmd_input = str(sequence)+"\n"+str(target)+"\n"
    os.system("echo '"+cmd_input+"'| RNAeval -v |grep Multi | tr -d A-Z-a-z-'('-')' | cut -d ':' -f 1 > multi.loop")
    
    multi = []
    with open("multi.loop", "r") as loops : 
        data_r = loops.read().split("\n")
        for dt in data_r : 
            if dt != '' : 
                multi.append(tuple(numpy.array(map(int, dt.strip().split(',')))-1))
        
    return multi 

def generatesequence(target, bp_pos, p_c, p_n) : 
    nucleotides = ["A", "C", "G", "U"]
    basepaires = ["AU", "UA", "CG", "GC", "GU", "UG"]
    
    
    arn = numpy.random.choice(nucleotides,len(target), p=p_n)

    for bp_cord in bp_pos : 
        bp = numpy.random.choice(basepaires,1, p=p_c)
        arn[bp_cord[0]] = bp[0][0]
        arn[bp_cord[1]] = bp[0][1]
        
    return "".join(arn)

p_c = [0.,0.,0.,1.,0.,0.]
p_n = [1.,0.,0.,0.]

loop_density = []
for elt in eterna_strc: 
    pos = get_bp_position(elt[0])
    seq = generatesequence(elt[0], pos, p_c, p_n)
    #print seq
    multi = getMultiCoord(seq,elt[0])
    #print multi 
    
    hairpins = getHairepinCoord(seq,elt[0])
    #print hairpins
    
    loop_density.append(float(len(multi)))

plt.scatter(lengths, loop_density,c="r" )
plt.title("Eterna dataset")
plt.xlabel("Length of target")
plt.ylabel("Loop_density")
plt.show()
```

```python
df = pd.read_csv("Rfam.csv")
#print df
rfam_lengths = []
rfam_loop_density = []
rfam_all_maker = []
rfam_C= [] 

for line in df.values : 
    rfam_lengths.append(len(line[-2]))
    pos = get_bp_position(line[-2])
    seq = generatesequence(line[-2], pos, p_c, p_n)
    #print seq
    multi = getMultiCoord(seq,line[-2])
    #print multi 
    
    hairpins = getHairepinCoord(seq,line[-2])
    #print hairpins
    
    rfam_loop_density.append(float(len(multi)+len(hairpins))/float(len(line[-2])))
    rfam_all_maker.append(line[-3])
    if line[-1] == "A": 
        rfam_C.append('blue')
    else : 
        if line[-1] == "B": 
            rfam_C.append('black')
        else : 
            rfam_C.append('orange')
plt.scatter(rfam_lengths,rfam_loop_density, c = rfam_C)
plt.show()
```

```python
def fold(sequence) : 
    rnafold = subprocess.Popen(args=["RNAfold", "-P", "rna_turner2004.par"], stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    out, err = rnafold.communicate(sequence)
    out = out.split()
    strc = out[1]
    mef = list(out[-1]) 
    for char in mef : 
        if char in ["(", ")", " "] : 
            mef.remove(char)
    
    return (strc,float("".join(mef)))

nemo_seqs = []
with open("nemo.seqs", "r") as f : 
    data = f.read()
    data = data.split("\n")
    #print data
    for i in range(0,len(data), 2) : 
        nemo_seqs.append(data[i])
    nemo_seqs.remove("")
nemo_eterna = []
with open("nemo.eterna", "r") as f : 
    structures = f.read()
    structures = structures.split()
    
    for strc in structures : 
        for i in range(len(eterna_strc)) : 
            if RNA.hamming_distance(strc,eterna_strc[i][0]) == 0.0 : 
                nemo_eterna.append(eterna_strc[i])
                
            
assert len(nemo_seqs) == len(structures)
    
assert [len(seq.strip()) for seq in nemo_seqs]==[len(strc.strip()) for strc in structures]

```

```python
def sort_result(success) : 
    i = 1
    sorted_success = []
    while i <= len(success)+1 :

        for entry in success : 
            if i== entry[-1] : 
                sorted_success.append(entry[0])
                break;

        i = i +1
    return sorted_success
```
