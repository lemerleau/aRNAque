## NEMO

```python
def fold(sequence) : 
    rnafold = subprocess.Popen(args=["RNAfold"], stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    out, err = rnafold.communicate(sequence)
    out = out.split()
    strc = out[1]
    mef = list(out[-1]) 
    for char in mef : 
        if char in ["(", ")", " "] : 
            mef.remove(char)
    
    return (strc,float("".join(mef)))

nemo_seqs = []
with open("../../data/eterna/nemo.seqs", "r") as f : 
    data = f.read()
    data = data.split("\n")
    #print data
    for i in range(0,len(data), 2) : 
        nemo_seqs.append(data[i])
    nemo_seqs.remove("")

```
